# DT Classifier

```python
#split dataset in features and target variable
feature_cols = ["CDRSB", "ADAS11", "ADAS13", "MMSE", "RAVLT_immediate", "RAVLT_learning", "RAVLT_forgetting","RAVLT_perc_forgetting", "MOCA_bl","FAQ_bl", "Entorhinal_bl", "Fusiform_bl", "Hippocampus_bl", "ICV_bl", "MidTemp_bl", "Ventricles_bl", "WholeBrain_bl", "AV45_bl", "FDG_bl", "ABETA_bl", "TAU_bl", "PTAU_bl"]
X = AD_bl[feature_cols] # Features
## I remeve "ABETA_bl" column becouse it contains >1700 ##################
```

```python
X
```

```python
y = AD_bl.DX_bl 
```

```python
y
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1) # 80% training and 20% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()
```

```python
# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
```

```python
#df[df.id.apply(lambda x: x.isnumeric())]
```

```python
#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# build and print the confusion matrix 
labels = [1,2,3]
cm = confusion_matrix(y_test, y_pred, labels=labels)
cm_df = pd.DataFrame(cm, index=labels, columns=labels)
print("Confusion Matrix:\n{}".format(cm_df))
```

```python
target_names=['CN','LMCI', 'AD']
print(classification_report(y_test,y_pred,target_names=target_names))
```
